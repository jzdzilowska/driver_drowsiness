# Takes videos from "uta-reallife-drowsiness-dataset" and extracts frame from every 3 seconds of each video for download as JPEG

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory
import cv2
import os

from glob import glob
# Importing all necessary libraries  
```

# Each cell has folder number part 1 and part 2

# To download each folder, go to save version - quick save (make sure it says "always save output"), the folder then can be found on the main page of the code file where you press edit

```python
#FOLDER 1 -- PARTS 1 AND 2

print("Folder 1 - Part 1")
if not os.path.exists('/kaggle/working/Fold1_part1'):
            os.mkdir('/kaggle/working/Fold1_part1')
            folder_name = '/kaggle/working/Fold1_part1/'

for dirname, _, filenames in os.walk('/kaggle/input/uta-reallife-drowsiness-dataset/Fold1_part1'):

    #Get Participant Number
    participant = dirname.split('/')[-1]

    #For each file in the folder
    for filename in filenames:
        #Get Path and Capture video using cv2
        pathway = os.path.join(dirname, filename)
        print(os.path.join(dirname, filename))
        cam = cv2.VideoCapture(pathway)

        #Get video Classification
        vid = filename[:-4]
        
        #Get frame number
        currentframe = 0
          
        while(cam.isOpened()): 
            #Reading from frame 
            ret,frame = cam.read() 
            
            if ret: 
                #If timestamp of frame is at 3 second mark
                if int(cam.get(0)) % 3000 == 0:
                    #If video is still left continue creating images 
                    #File Name Format -- Participant_#_Vid_#_Frame_#.jpg
                    name = folder_name + 'Participant_' + str(participant) + "_Vid_" + str(vid) + '_Frame_' + str(currentframe) + '.jpg'
              
                    #Writing the extracted images 
                    cv2.imwrite(name, frame) 

                    #Increase Frame Number
                    currentframe += 1
            else: 
                break
        cam.release()  

print("Folder 1 - Part 2")
if not os.path.exists('/kaggle/working/Fold1_part2'):
            os.mkdir('/kaggle/working/Fold1_part2')
            folder_name = '/kaggle/working/Fold1_part2/'

for dirname, _, filenames in os.walk('/kaggle/input/uta-reallife-drowsiness-dataset/Fold1_part2'):

    #Get Participant Number
    participant = dirname.split('/')[-1]

    #For each file in the folder
    for filename in filenames:
        #Get Path and Capture video using cv2
        pathway = os.path.join(dirname, filename)
        print(os.path.join(dirname, filename))
        cam = cv2.VideoCapture(pathway)

        #Get video Classification
        vid = filename[:-4]
        
        #Get frame number
        currentframe = 0
          
        while(cam.isOpened()): 
            #Reading from frame 
            ret,frame = cam.read() 
            
            if ret: 
                #If timestamp of frame is at 3 second mark
                if int(cam.get(0)) % 3000 == 0:
                    #If video is still left continue creating images 
                    #File Name Format -- Participant_#_Vid_#_Frame_#.jpg
                    name = folder_name + 'Participant_' + str(participant) + "_Vid_" + str(vid) + '_Frame_' + str(currentframe) + '.jpg'
              
                    #Writing the extracted images 
                    cv2.imwrite(name, frame) 

                    #Increase Frame Number
                    currentframe += 1
            else: 
                break
        cam.release()
```
